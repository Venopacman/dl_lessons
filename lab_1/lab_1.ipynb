# Часть вторая. Custom CNN
Обучим свёрточную нейронную сеть и оценим ошибку.

Для работы с нейронными сетями мною был выбран PyTorch, в виду хорошей документации и python-like API

```python
import torch.nn as nn
import torch.nn.functional as F
```

Архитектура бейзлайн сети -- модификация сети из официального туториала, были добавлены zero-paddings, дополнительный сверточный слой + увеличено общее количество каналов в свертках. Размер Батча для обучения = 6, тк на больших значениях сеть обучалась очень медленно

```python
class BaselineCNN(nn.Module):
    def __init__(self):
        super(BaselineCNN, self).__init__()
        self.conv1 = nn.Conv2d(3, 6, 5, padding=1)
        self.pool = nn.MaxPool2d(2, 2)
        self.conv2 = nn.Conv2d(6, 12, 5, padding=1)
        self.conv3 = nn.Conv2d(12, 24, 5, padding=1)
        self.fc1 = nn.Linear(16224, 400)
        self.fc2 = nn.Linear(400, 50)
        self.fc3 = nn.Linear(50, 2)

    def forward(self, x):
        x = self.pool(F.relu(self.conv1(x)))
        x = self.pool(F.relu(self.conv2(x)))
        x = self.pool(F.relu(self.conv3(x)))
        x = x.view(-1, 16224)
        x = F.relu(self.fc1(x))
        x = F.relu(self.fc2(x))
        x = F.softmax(self.fc3(x), dim=1)
        return x


net = BaselineCNN()
```

```python
trainloader = torch.utils.data.DataLoader(trainset, shuffle=True, batch_size=6)
valloader = torch.utils.data.DataLoader(valset, shuffle=False, batch_size=6)
```

```python
def get_accuracy(network, loader):
    network = network.train(False)
    accuracy = 0.0
    for img, label in loader:
        label_hat = torch.argmax(network(img),dim=1).numpy()
        accuracy += np.sum(label.numpy() == label_hat)
    accuracy /= loader.dataset.indices.size()[0]
    network = network.train(True)
    return accuracy
#         print(label.numpy()[0]==label_hat[0])
```

```python
import torch.optim as optim

criterion = nn.CrossEntropyLoss()
optimizer = optim.SGD(net.parameters(),lr=3e-2, momentum=0.8)
```

```python
%%time
net = torch.load("best_baseline_cnn.pt")
best_val_accuracy = get_accuracy(net, valloader)
print(best_val_accuracy)
```

```python
optimizer = optim.SGD(net.parameters(),lr=3e-4, momentum=0.7)
```

Обучаем сеть стохастическим градиентным спуском + метод моментов + мини-батч, со временем уменьшая learning rate в 2 раза, для улучшения процесса сходимости. 

На каждой эпохе проверяем точность на валидационной выборке и сохраняем модель с если точность улучшилась, таким образом пытаемся избежать переобучения

```python
lr = 3e-4
for epoch in tqdm(range(20)):  # loop over the dataset multiple times
    
    if epoch % 5 == 4:
        lr/=2
        optimizer.param_groups[0]['lr']=lr
    running_loss = 0.0
    for inputs, labels in trainloader:

        # zero the parameter gradients
        optimizer.zero_grad()
        # forward + backward + optimize
        outputs = net(inputs)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()
        # print statistics
        running_loss += loss.item()
        
    loss = running_loss / len(trainloader)
    val_accuracy = get_accuracy(net, valloader)
    if val_accuracy > best_val_accuracy:
        best_val_accuracy = val_accuracy
        print("New best validation accuracy: {0} !".format(best_val_accuracy))
        torch.save(net, "best_baseline_cnn.pt")
    print('Epoch {0}, loss: {1}, learning rate:{2}'.format(epoch + 1, loss, lr))
    running_loss = 0.0
        
# torch.save(net, "my_best_ann.pkl")
print('Finished Training')
```

```python
net = torch.load("best_baseline_cnn.pt")
val_accuracy = get_accuracy(net, valloader)
val_accuracy
```

В итоге получили точность в районе 85%, при том что архитектура сети была довольно простой и есть возможность её улучшить

### Data augmentation

Будем случайно изменять цветовые характеристики изображений, а также поворачивать их под случайным углом

```python
augment_transform = transforms.Compose([
        transforms.ColorJitter(brightness=0.3, contrast=0.3, saturation=0.3, hue=0.1),
        transforms.RandomAffine(45),
        transforms.Resize(224),
        transforms.CenterCrop(224),
        transforms.ToTensor()
        ])
transform_test = transforms.Compose([
    transforms.Resize(224),
    transforms.CenterCrop(224),
    transforms.ToTensor()])

trainset = torchvision.datasets.ImageFolder("../dl_hotdog_data/train", transform = augment_transform)
val_size = int(len(trainset.imgs)*0.2)
train_size = len(trainset.imgs) - val_size 

trainset = torchvision.datasets.ImageFolder("../dl_hotdog_data/train", transform = augment_transform)
val_size = int(len(trainset.imgs)*0.2)

trainset, valset = torch.utils.data.dataset.random_split(trainset, [train_size, val_size]) 

valset.dataset.transform = transform_test

testset = torchvision.datasets.ImageFolder("../dl_hotdog_data/test", transform=transform_test)
testloader = torch.utils.data.DataLoader(testset)
```

```python
trainloader = torch.utils.data.DataLoader(trainset, batch_size=6, shuffle=True)
valloader = torch.utils.data.DataLoader(valset)
```

### Vgg16 fine-tuning

Для дообучения модели опять снимем последний выходной слой, поменяем его с 1к нейронов на 2 и попробуем дообучить на наших аугментированных данных, постепенно уменьшая скорость обучения

```python
from torchvision.models import vgg16
vgg_model = vgg16(True)
for param in vgg_model.parameters():
    param.requires_grad = False
    
vgg_model.classifier[-1] = nn.Linear(4096,2)

for child in vgg_model.classifier:
    for param in child.parameters():
        param.requires_grad = True
```

```python
import torch.optim as optim

criterion = nn.CrossEntropyLoss()
optimizer = optim.SGD(vgg_model.classifier.parameters(), lr=5e-4, momentum=0.9)
```

```python
best_val_accuracy = 0
```

```python
lr = 5e-4
for epoch in tqdm(range(3)):  # loop over the dataset multiple times
    running_loss = 0.0
    for inputs, labels in tqdm(trainloader):

        # zero the parameter gradients
        optimizer.zero_grad()
        # forward + backward + optimize
        outputs = F.softmax(vgg_model(inputs), dim=1)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()
        # print statistics
        running_loss += loss.item()
        
    loss = running_loss / len(trainloader)
    val_accuracy = get_accuracy(vgg_model, valloader)
    if val_accuracy > best_val_accuracy:
        best_val_accuracy = val_accuracy
        print("New best validation accuracy: {0} !".format(best_val_accuracy))
        torch.save(vgg_model, "best_vgg.pt")
    print('Epoch {0}, loss: {1}, learning rate:{2}'.format(epoch + 1, loss, lr))
    running_loss = 0.0
    lr/=2
    optimizer.param_groups[0]['lr']=lr
        
print('Finished Training')
```
